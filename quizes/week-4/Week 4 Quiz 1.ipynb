```python
# week 4 quiz 1 question 4
from itertools import combinations
from pprint import pprint as pp

def generate_input(n, m):
    attributes = [str(i) + str(j)
                  for i in xrange(1, n + 1)
                  for j in xrange(1, m + 1)]
    objects = [str(i) + str(j)
               for i in xrange(1, n + 1)
               for j in xrange(1, m + 1)]


    objs_by_attr = {attr: set() for attr in attributes}
    attrs_by_obj = {obj: set() for obj in objects}

    for obj in objects:
        for attr in attributes:
            if obj[0] <= attr[0]:
                objs_by_attr[attr].add(obj)
                attrs_by_obj[obj].add(attr)
    
    attributes_set = set(attributes)
    objects_set = set(objects)
    
    return attributes, attributes_set, objs_by_attr, objects, objects_set, attrs_by_obj
    

def obj_closure(objects):
    m = set(attributes)
    for obj in objects:
        m = m.intersection(attrs_by_obj[obj])
    return m

def attr_closure(attributes):
    g = set(objects)
    for attr in attributes:
        g = g.intersection(objs_by_attr[attr])
    return g
    
def closure(s):
    return obj_closure(attr_closure(s))


def get_pseudo_intents():
    pseudo_intents = list()
    
    for i in xrange(0, len(attributes_set) + 1):
        for subset in list(combinations(attributes_set, i)):
            subset = set(subset)
            if subset != closure(subset):
                is_pseudo_intent = True
                for pseudo_intent in pseudo_intents:
                    if pseudo_intent < subset:
                        if not closure(pseudo_intent) <= subset:
                            is_pseudo_intent = False
                            break
                if is_pseudo_intent:
                    pseudo_intents.append(subset)
    return pseudo_intents


def get_preclosed_intents():
    pseudo_intents = list()
    
    for i in xrange(0, len(attributes_set) + 1):
        for subset in list(combinations(attributes_set, i)):
            subset = set(subset)
            is_pseudo_intent = True
            for pseudo_intent in pseudo_intents:
                if pseudo_intent < subset:
                    if not closure(pseudo_intent) <= subset:
                        is_pseudo_intent = False
                        break
            if is_pseudo_intent:
                pseudo_intents.append(subset)
    return pseudo_intents


n = 3
m = 2
attributes, attributes_set, objs_by_attr, objects, objects_set, attrs_by_obj = generate_input(n, m)
pseudo_intents = get_pseudo_intents()

pp(len(pseudo_intents))
pp(pseudo_intents)
```

```python
# week 4 quiz 1 question 8
attrs_by_obj = {
    '1': set(['b', 'c']),
    '2': set(['a', 'b', 'c']),
    '3': set(['a', 'c', 'd']),
    '4': set(['b', 'd']),
    }

attributes, attributes_set, objs_by_attr, objects, objects_set, attrs_by_obj = generate_input(attrs_by_obj)

pseudo_intents = get_pseudo_intents()

pp('Pseudo intents:')
pp(len(pseudo_intents))
pp(pseudo_intents)

def preclosure_operator(intent_set, pseudo_intents):
    prev_intents = None
    cur_intents = intent_set
    while prev_intents != cur_intents:
        pseudo_intents_union = set()
        for pseudo_intent in pseudo_intents:
            if pseudo_intent < cur_intents:
                pseudo_intents_union = pseudo_intents_union.union(closure(pseudo_intent))
        prev_intents = cur_intents
        cur_intents = cur_intents.union(pseudo_intents_union)
    return cur_intents
    

r1 = preclosure_operator({'a', 'b'}, pseudo_intents)
r2 = preclosure_operator({'a', 'd'}, pseudo_intents)
pp(r1)
pp(r2)
pp('Answer is:')
pp(r1.intersection(r2))
```
